## Earning patterns over its theater lifespan

```python
movies_df
```

```python
movies_df.loc[movies_df['Movie_Title']== 'Jurassic World (2015)']
```

```python
movies_df_jurassic = movies_df.loc[movies_df['Movie_Title']== 'Jurassic World (2015)'].copy()
```

```python
movies_df_jurassic['Date'] = pd.to_datetime (movies_df_jurassic['Date'])
```

```python
fig = plt.figure(figsize=(12,6))

x = movies_df_jurassic['Date']
y = movies_df_jurassic['Daily']

chart = plt.plot(x,y)

#plt.plot(linewidth=2)

plt.title("Jurassic World Earnings Over Time")
plt.xlabel("Timespan")
plt.ylabel("Earnings (Tens of Millions)")

plt.tight_layout()
```

```python
plt.grid()
```

### Earning are at their highest at movie opening. Earnings have a study decline with weekend spikes.
